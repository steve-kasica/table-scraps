This file contains lists of all code repositories collected during our the repo selection phase of our technical observation study.

1. **Pool of Journalistic Repositories** The first pool of repos constitute journalistic repos, gather from a systematic search of [GitHub](https://github.com) and [ObservableHQ](https://observablehq.com). 

1. **Pool of Journalistic Repositories with Data Analysis**
The second list constitutes our pool of journalistic repositories containing data analysis, which was obtained from manually inspecting the contents of the first pool. We include this second list as part of our secondary contributions.

```python
import pandas as pd
from IPython.core.display import Markdown

def displayList(df):
    """Display repo list as an HTML description details element."""
    out = ''
    for i, row in df.iterrows():
        out += """
<dt><a href="{url}">{title}</a></dt>
<dd><strong>News Organization</strong>: <em>{org}</em></dd>
<br>
        """.format(**{
            'title': row['url'][19:],
            'url': row['url'],
#            'description': row['description'],
            'org': row['org'],
#            'created_at': row['created_at']
        })

    return Markdown('<dl>{}</dl>'.format(out))

# Load repos
repos = pd.read_csv('data/repos.csv')

#dateparse = lambda x: str(x)[0:4]
#repos['created_at'] = repos['created_at'].apply(dateparse).replace('nan', 'Not Available')
repos = repos.fillna('Not Available')
```

# Pool of Journalistic Repos

```python
txt = """
Below is an index into {} journalistic repositories we located through a systematic search of [github.com](https://github.com/) and [ObservableHQ](https://observablehq.com/).
"""
Markdown(txt.format(repos.shape[0]))
```
